## Lets start by importing image dataset inbuilt within Scikit-learn. Also lets understand the structure and its shape!

```python
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces()
print faces.keys()
print faces.images.shape
print faces.data.shape
print faces.target.shape
```

## Lets understand the max, min and mean of the input data. This will help us to decide whether we have to normalize the data.

```python
print np.max(faces.data)
print np.min(faces.data)
print np.mean(faces.data)
```

## Per above results, since it falls in the range of 0 to 1, there is no need to normalize!

## Defining helper function to display images from datasets given the image array and no. of images to show.

```python
def print_faces(images, target, top_n):
    fig = plt.figure(figsize=(12,12))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
    
    for i in range(top_n):
        p = fig.add_subplot(20, 20, i+1, xticks=[], yticks=[])
        p.imshow(images[i], cmap=plt.cm.bone)
        p.text(0, 14, str(target[i]))
        p.text(0, 60, str(i))
        
```

```python
%pylab inline
print_faces(faces.images, faces.target, 20)
```

## Now lets train the Support Vector Machine

```python
from sklearn.svm import SVC
svcClassifier = SVC(kernel = 'linear')
```

## Split dataset into training and test

```python
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test =train_test_split(faces.data, faces.target, test_size=0.25, random_state=0)
```

## Function to evaluate K-fold cross-validation

```python
from sklearn.cross_validation import cross_val_score, KFold
from scipy.stats import sem

def evaluate_cross_validation(classifier, x_train, y_train, numberOfKFolds):
    cv = KFold(len(y_train), numberOfKFolds, shuffle=True, random_state=0)
    scores = cross_val_score(classifier, x_train, y_train, cv=cv)
    print scores
    print ("mean score: {0:.3f} (+/- {1:3f})").format(np.mean(scores), sem(scores))
```

## Evaluate the cross validation

```python
evaluate_cross_validation(svcClassifier, x_train, y_train, 5)
```

## Now lets perform training on training set and evaluate its performance on testing set

```python
from sklearn import metrics

def train_and_evaluate(svcClassifier, x_train, x_test, y_train, y_test):
    svcClassifier.fit(x_train, y_train)
    print "Accuracy on training set:"
    print svcClassifier.score(x_train, y_train)
    print "Accuracy on testing set:"
    print svcClassifier.score(x_test, y_test)
    
    y_pred = svcClassifier.predict(x_test)
    
    print "Classification Report:"
    print metrics.classification_report(y_test, y_pred)
    print "Confusion Matrix:"
    print metrics.confusion_matrix(y_test, y_pred)
```

## Now lets evaluate!

```python
train_and_evaluate(svcClassifier, x_train, x_test, y_train, y_test)
```

## Now lets classify people with and without glasses

```python
glasses = [(10, 19), (30, 32), (37, 38), (50, 59), (63, 64), (69, 69), (120, 121), (124, 129), (130, 139), (160, 161), (164, 169), (180, 182), (185, 185), (189, 189), (190, 192), (194, 194), (196, 199), (260, 269), (270, 279), (300, 309), (330, 339), (358, 359), (360, 369)]
```

## Now lets define a function based on the above segments returns new target array that marks 1 for faces with glasses and 0 for faces without glasses.

```python
def create_target(segments):
    y= np.zeros(faces.target.shape[0])
    for(start, end) in segments:
        y[start:end + 1] = 1
    return y
```

```python
target_glasses = create_target(glasses)
```

## Perform training/testing split again with new dataset

```python
x_train, x_test, y_train, y_test = train_test_split(faces.data, target_glasses, test_size=0.25, random_state=0)
```

## Create new SVC classifier

```python
svc_2 = SVC(kernel='linear')
```

```python
evaluate_cross_validation(svc_2, x_train, y_train, 5)
```

```python
train_and_evaluate(svc_2, x_train, x_test, y_train, y_test)
```

```python
x_test = faces.data[30:40]
y_test = target_glasses[30:40]

#Fill with ones given a size
select = np.ones(target_glasses.shape[0])
#Fill [30:40] with zeros
select[30:40] = 0

#print select

x_train = faces.data[select == 1]
y_train = target_glasses[select == 1]

#print x_train
#print y_train

svc_3 = SVC(kernel='linear')
train_and_evaluate(svc_3, x_train, x_test, y_train, y_test)

```

## And finally lets see the prediction results!

```python
y_pred = svc_3.predict(x_test)
eval_faces = [np.reshape(a, (64, 64)) for a in x_test]
print_faces(eval_faces, y_pred, 10)
```
